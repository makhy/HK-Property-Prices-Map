```python
import os
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import webbrowser
from IPython.display import HTML, display
import branca
```

```python
#https://towardsdatascience.com/data-visualization-with-python-folium-maps-a74231de9ef7
latitude = 22.312287
longitude = 114.181424
map = folium.Map(location=[latitude, longitude], zoom_start=12.5)    
```

```python
df0 = pd.read_csv("makgeo_updated.csv", index_col = 0)
df = df0.groupby("area").mean()
```

```python
df
```

```python
df['sell_quartile'] = pd.qcut(df['sell_price'], 4, labels=False)
```

```python
df.columns
```

```python
df.reset_index(level=0, inplace=True)
```

```python
# df["lat"].iloc[0] = 22.319945
# df["long"].iloc[0] = 114.168728

# df["lat"].iloc[7] =22.318098
# df["long"].iloc[7] =114.160499

# df["lat"].iloc[8] =22.325893
# df["long"].iloc[8] =114.177764

# df["lat"].iloc[9] =22.220680
# df["long"].iloc[9] =114.213372

# df["lat"].iloc[13] =22.333023
# df["long"].iloc[13] =114.178874
```

```python
df = df.fillna('')
```

```python
df.head()
```

```python
colordict = {0: 'blue', 1: 'green', 2: 'orange', 3: 'red'}
```

```python
# import folium
# from IPython.display import display
# LDN_COORDINATES = (22.312287, 114.181424)
# myMap = folium.Map(location=LDN_COORDINATES, zoom_start=12)
# display(myMap)
```

```python
for lat, lon, sq, sp, rent, city in zip(df['lat'], df['long'], df['sell_quartile'], df['sell_price'], df['rent_price'], df['area']):
    folium.CircleMarker(
        [lat, lon],
        radius=.00026*rent,
        popup = ('City: ' + str(city).capitalize() + '<br>'
                 'Rent: ' + str(rent) + '<br>'
                 'Selling Price: ' + str(sp) +'$'
                ),
        color='b',
        key_on = sq,
        threshold_scale=[0,1,2,3],
        fill_color = colordict[sq],
        fill=True,
        fill_opacity=0.8
        ).add_to(map)

#create legend
#https://medium.com/@bobhaffner/creating-a-legend-for-a-folium-map-c1e0ffc34373    
    
legend_html =   '''
                <div style="position: fixed; 
                            bottom: 50px; left: 50px; width: 100px; height: 140px; 
                            border:4px solid black; z-index:9999; font-size:12px;
                            ">&nbsp; Property Prices <br>
                              &nbsp; Sell$ 0Q &nbsp; <i class="fa fa-map-marker fa-2x" style="color:blue"></i><br>
                              &nbsp; Sell$ 25Q &nbsp; <i class="fa fa-map-marker fa-2x" style="color:green"></i><br>
                              &nbsp; Sell$ 50Q &nbsp; <i class="fa fa-map-marker fa-2x" style="color:orange"></i><br>
                              &nbsp; Sell$ 75Q &nbsp; <i class="fa fa-map-marker fa-2x" style="color:red"></i>
                </div>
                ''' 

map.get_root().html.add_child(folium.Element(legend_html))

map
```
