```python
from bs4 import BeautifulSoup
import googlemaps
import requests  
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("./group4_hkprop_updated.csv",sep=",")
df = df.drop(columns = ["Unnamed: 0"])
```

```python
df.head(2)
```

```python
gmaps = googlemaps.Client(key = "YOUR KEY")
```

```python
# geocoder.google('some address', key='YOUR-API-KEY-HERE')
# table.loc[:,'Total'] = table.sum(axis=1)
# table.head()
# df['lat'] = geocoder.google(df[16]).latlng[0]
#df['lng'] = geocoder.google(df[16]).latlng[1]

list = df.area.unique()
```

```python
# https://www.youtube.com/watch?v=27RkZjVmarM
for place in list:
    single_place = place.split("/")[0]
    try:
        geocode = gmaps.geocode(single_place)[0]
        latlong = geocode["geometry"]["location"]
    except IndexError:
        latlong = {"lat": None,
                  "lng": None}
    df.loc[df.area == place, "lat"] = latlong["lat"]
    df.loc[df.area == place, "long"] = latlong["lng"]
```

```python
df.head(1)
```

```python
df.groupby("area").head()
```

```python
df["sell_price"] = df["sell_price"].str.extract("(\d+\.\d+)")
```

```python
df["sell_price"] = df["sell_price"].str.replace(".", "")
```

```python
df["rent_price"] = df["rent_price"].str.extract("(\d+\.\d+)")
```

```python
df["rent_price"] = df["rent_price"].str.replace(".", "")
```

```python
df['sell_price'] = df['sell_price'].astype(int)
df['rent_price'] = df['rent_price'].astype(int)
```

```python
df.to_csv("makgeo_updated.csv")
```
